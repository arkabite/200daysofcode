# **Correlation and Covariance:**

#### **1.Covariance:**

Covariance tell you how two variable are related, doi they increase/decrease together.

- +ve->Move in the same direction
- -ve->Move in the opposite direction
- Near 0-> little relationship
<br> 
<br>It compares the deviation of each variables mean.

Problem: Magnitude depends on units which can be hard to interpret

```python
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

cov_matrix = np.cov(x, y)

print("Covariance:", cov_matrix)
print("Covariance:", cov_matrix[0, 1])
```

#### **2.Correlation:**

Correlation is standardized covariance , basically indicating how strong the relationship is between the varaibles.

Range: -1<=r<=1

- +1 indicates a perfect positive linear relationship.
- -1 indicates a perfect negative linear relationship
- 0 indicates no linear relationship.
<br> 

```python
import pandas as pd

corr = np.corrcoef(x, y)[0, 1]
print("Pearson correlation:", corr)
df = pd.DataFrame({"x": x, "y": y})
print("Spearman correlation:\n", df.corr(method="spearman"))
```

**Correlation ≠ Causation:**

Just because variables move together does not mean one causes the other.

Eg:

- Ice cream sales ↑
- Drowning incidents ↑

Cause? → hot weather.

Correlation only shows association, not mechanism.

**Multicollinearity:**

Occurs when features are highly correlated with each other.

Eg:

House size in sqft and House size in square meters both convey same information.

_Multicollinearity causes:_

❌ unstable model coefficients
❌ unreliable interpretation
❌ inflated variance
❌ poor generalization

Especially harmful in linear models.

```python
import pandas as pd

data = pd.DataFrame({
    "sqft": [1000, 1200, 1500, 1800],
    "sqm": [93, 111, 139, 167]  # highly correlated
})

print(data.corr())
```
