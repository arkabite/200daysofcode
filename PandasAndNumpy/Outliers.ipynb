# **Outliers:**

Data that sits far away from the rest off the data, it isnt necessarily always bad and can represent edge cases.

#### **1.When to remove vs when to keep an outlier:**

**Remove when:**
- Data entry error
- Impossible value
- Instrument failure

**Example:**
Age = 500 → clearly invalid.

**Keep when:**

- Represents rare but real event
- Important edge case
- Business-critical anomaly

**Example:**
Fraud detection → extreme transactions matter.

#### **2.Impact on regression models:**

Regression tries to fit a line minimizing error.

Outliers can:   
- Pull the regression line
- Inflate error
- Distort coefficients
- Reduce prediction accuracy

Especially harmful in linear regression.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 50])  # outlier

model = LinearRegression()
model.fit(X, y)

print("Slope:", model.coef_)
print("Intercept:", model.intercept_)
```
