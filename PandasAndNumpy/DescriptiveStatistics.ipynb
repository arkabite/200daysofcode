# **Descriptive Statistics:**

#### **1.Mean vs Median:**

`Mean(average):` sum of all values/number of values
<br>

<br>`Median` is the obtained middle value after sorting

**Q) When is mean misleading?**
<br>->Mean is sensitive to Outliers

```python
[5, 6, 7, 8, 100]
# Mean is pulled upwards
# Median still represents centre
```

```python
import numpy as np
import pandas as pd

arr=np.array([5, 6, 7, 8, 100])
print("Mean:", np.mean(arr))    
print("Median:", np.median(arr))
```

#### **4.Variance and Standard Deviation:**

They measure spread.

`Variance` is the average squared distance from the mean.
<br> 
<br>`Standard Deviation` is the square root of variance.
<br> 

- Higher value idicates more spread
- Lower value indicated tightly clustered data

```python
arr=np.array([5, 6, 7, 8, 9])

print("standard deviation:", np.std(arr))
print("variance:", np.var(arr))
```

#### **3.Range and IQR:**

`Range` is a simple spread measure,

                                    Range=max(x)-min(x)                              

`IQR (Inter Quartile range)` measures the spread of the middle 50%,

                                    IQR=Q3-Q1                                        

<br>**Q1**->25th percentile
<br>**Q**->50th percentile
<br>**Q3**->75th percentile

<br>**Note:** IQR is less sensitive to outliers and more robust

```python
arr=np.array([5, 6, 7, 8, 9])

print("Range:",max(arr)-min(arr))

IQR=np.percentile(arr,75)-np.percentile(arr,25)
print("Interquartile Range:", IQR)

print("25th percentile:", np.percentile(arr, 25))
print("Median:", np.percentile(arr, 50))
print("75th percentile:", np.percentile(arr, 75))
```

#### **4.Outlier:**

Outliers are Extreme values that distort statistics.

`IQR Method`:
- Outlier boundaries
<br> 

<br>`Z-score Intuition`:
- Measures distance from mean in Standard Deviation.
<br> 

```python
data = np.array([5, 7, 8, 12, 15, 20, 100])

Q1=np.percentile(data, 25)
Q3=np.percentile(data, 75)
lower=Q1-1.5*(Q3-Q1)
upper=Q3+1.5*(Q3-Q1)
outliers=[x for x in data if x<lower or x>upper]
print("Outliers:", outliers)

mean=np.mean(data)
std=np.std(data)
z_scores=[(x-mean)/std for x in data]
print("Z-scores:", z_scores)                                                                                           
```

#### **5.Skewness:**

Describes distribution shape

`Right Skew (Positive Skew)`:
- long tail on the right
- Mean > Median
Example: Income data

`Left Skew (Negative Skew)`:
- Tail on left
- mean < median

```python
data = pd.Series([5, 6, 7, 8, 100])

print("Skewness:", data.skew())
```

Interpretation:

- Skew ≈ 0 → symmetric

- Skew > 0 → right skew

- Skew < 0 → left skew
