# **Probability and Distributions:**

#### **1.What is a distribution?**

A distribution describes:
- How a value is spread
- How often they occur
- The shape of the data

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randint(1, 10, 1000)

plt.hist(data)
plt.title("Example Distribution")
plt.show()
```

#### **2.Normal Distribution:**

Also known as the `bell curve`.

**Properties:**
- Symmetric
- Mean=Median=Mode
- Most values near center
- Predictable speed

```python
data = np.random.normal(loc=50, scale=10, size=1000)

plt.hist(data, bins=30)
plt.title("Normal Distribution")
plt.show()
```

#### **3.Skewed Distributions:**

Not symmetric.

**Right skew (positive skew)**
- Long tail to the right
- Few extreme large values
- Mean pulled upward

Example: income

**Left skew (negative skew)**
- Tail on the left
- Few extreme small values

_Skew matters because Skewed data can make:_

- mean misleading
- models unstable

Median or transformations may work better.

```python
data = np.random.exponential(scale=2, size=1000)

plt.hist(data, bins=30)
plt.title("Right Skewed Distribution")
plt.show()
```

#### **4.Population vs Sample:**

`Population` is all the data we carte about whereas `Sample` is a subset used to **estimate** the population

Sampling is usually cheaper,faster and more practical.

```python
population = np.random.normal(50, 10, 100000)
sample = np.random.choice(population, size=500)

print("Population mean:", np.mean(population))
print("Sample mean:", np.mean(sample))
```

#### **5.Law of Large Numbers:**

As sample size increases:

- sample average â†’ population average

Small samples are noisy.
Large samples stabilize.

Ex:

Flipping a coin 10 times vs 10,000 times.

```python
means = []

for n in range(10, 1000, 50):
    sample = np.random.normal(50, 10, n)
    means.append(np.mean(sample))

plt.plot(means)
plt.title("Law of Large Numbers")
plt.show()
```

#### **6.Central Limit Theorem:**

Averages of samples tend to form a normal distribution â€” even if the original data is not normal.

Implication:

- Large samples behave predictably
- Enables statistical inference

```python
means = []

for _ in range(1000):
    sample = np.random.exponential(scale=2, size=30)
    means.append(np.mean(sample))

plt.hist(means, bins=30)
plt.title("CLT in Action")
plt.show()
```

Even in `skewed` data the sample means look normal.
