# **Types off Data**

#### **1.Numerical vs Categorical data:**

`Numerical` data signifies counts/quantites you can do Arithematic with.
<br>Eg: Age,Temperatures,Salary,etc;

`Categorical` data signifies labels or groups.
<br>Eg: Color,City,Product Type,etc;

```python
import pandas as pd

data=pd.DataFrame({
    'Name':['Alice','Bob','Charlie'],
    'Age':[25,30,35],
    'Salary':[50000,60000,70000]
})

print(data.dtypes)
```

#### **2.Discrete vs Continous data:**

These are sub types of **Numerical data.**

_`Discrete data:`_ Data that can be counted (i.e does not have decimal values), like:
- Number of cars
- Number of students

_`Countinous data:`_ Data that usually represents measures and can contain decimals, like:
- Height
- Weight
- Time

```python
Cars=[5,12,7,3,9] #Discrete data
Temperature=[22.5, 23.0, 21.8, 24.1, 22.9] #Continuous data
```

#### **3.Nominal vs Ordinal data:**

These are sub types of categorical data.

_`Nominal`_: Categorical data that does not have any natural order, like:
- Country
- Eye color
- Gender

_`Ordinal`_: Categorical data that does have a meaningful/natural order, like:
- Sizes (small/medium/large)
- Educational level
- Ratings (1-5 stars)

```python
Country=['USA', 'India', 'UK', 'Germany', 'France'] #Nominal data
Education_Level=['High School', 'Bachelor', 'Master', 'PhD', 'Bachelor'] #Ordinal data
```

#### **4.Binary variables:**

A special character variable with only two values, like:
- Yes/No
- Pass/Fail
- 0/1

```python
val=[0,1,0,1,0] #Binary data
```

#### **5.Target vs Features:**

`Features` are **_input variables that are used to make predictions_** whereas `Target` is the value that **_we try to predict_**.
<br> Eg: 
- Features->Location,Size,Rooms,etc;
- Target->Price
        

```python
data=pd.DataFrame({
    'size':[1000,1500,2000],
    'price':[200000,300000,400000],
    'location':['city','suburb','rural'],
    'rooms':[3,4,5],
})

features=data[['size','location','rooms']]
target=data['price']

print("Features:\n",features)
print("\nTarget:\n",target)
```
